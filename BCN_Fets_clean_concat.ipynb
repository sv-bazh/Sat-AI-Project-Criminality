```python
df2019 = pd.read_excel("Fets_2019.xls")
headers = df2019.iloc[2]
new_df2019  = pd.DataFrame(df2019.values[3:], columns=headers)
new_df2019.head()
```

```python
# concatenating all the dataframes for each year into a single one

frames = [new_df2011,new_df2012,new_df2013,new_df2014,new_df2015,new_df2016,new_df2017,new_df2018,new_df2019]
df_all = pd.concat(frames)
```

```python
df_all.head()
```

```python
df_all.info()
```

```python
list(df_all.columns)
```

```python
df_all["Àrea Bàsica Policial (ABP)"].unique()
```

```python
#deleting the "ABP" before each Area name

df_all['Àrea Bàsica Policial (ABP)'] = df_all['Àrea Bàsica Policial (ABP)'].str[4:]
```

```python
df_all["Àrea Bàsica Policial (ABP)"].unique()
```

```python
df_all["Regió Policial (RP)"].unique()
```

```python
#deleting the "RP" before each region name


df_all["Regió Policial (RP)"] = df_all["Regió Policial (RP)"].str[3:]
```

```python
df_all["Regió Policial (RP)"].unique()
```

```python
df_all["Mes"].unique()
```

```python
df_all["Nom mes"].unique()
```

```python
# capitalizing the first letter of month name

df_all["Nom mes"] = df_all["Nom mes"].str.capitalize()
```

```python
df_all["Nom mes"].unique()
```

```python
#the following lines are to add a Date column 

df_all["day"] = "28"
df_all.rename(columns={'Mes':'month'}, inplace=True)
df_all.rename(columns={'Any':'year'}, inplace=True)
df_all["date"] = pd.to_datetime(df_all[['year', 'month', "day"]])
df_all.head()
```

```python
df_all.to_csv( "fets_all.csv", index=False, encoding='utf-8-sig')
```
